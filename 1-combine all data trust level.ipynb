```python
import pandas as pd
import os
folder_path = r"/Users/e410675/Library/CloudStorage/OneDrive-LondonSouthBankUniversity/LSBU - Jose's Collaborative Projects/Norfolk and Waveney CHPPD/2025 November/Data Engineering/Data/Trust level"
combined_data = []

for file_name in os.listdir(folder_path):
    if file_name.endswith('.xlsx') or file_name.endswith('.xls'):
        file_path = os.path.join(folder_path, file_name)
        data = pd.read_excel(file_path)
        data['source file name'] = file_name  # Adding a column with the file name
        combined_data.append(data)

combined_data = pd.concat(combined_data, axis=0)
#combined_data.to_excel('NHS England trust level CHPPD - data combined.xlsx', index=False)
```

```python
import re
import pandas as pd

# --- 1. Função para extrair o padrão YYYY-MM da coluna 'source file name'
def extract_year_and_month(text):
    # Regex para encontrar datas no formato 2021-01
    pattern = r'\b(20\d{2}-\d{2})\b'
    match = re.search(pattern, text)
    if match:
        return match.group(1)
    else:
        return None

# --- 2. Aplicar a função para criar a coluna 'year_month'
combined_data['year_month'] = combined_data['source file name'].apply(extract_year_and_month)

# --- 3. Converter para datetime (formato ano-mês)
combined_data['year_month'] = pd.to_datetime(
    combined_data['year_month'], 
    format='%Y-%m'
).dt.to_period('M')

# --- 4. Verificação final
print(combined_data)
```

```python
# Combine columns with different names into one final column (without renaming first)

import pandas as pd

# dicionário: nome_final -> lista de nomes possíveis no df (incluindo os errados)
column_groups = {
    "CHPPD – Registered Nurses and Midwives": [
        "CHPPD – Registered Nurses and Midwives",
        "CHPPD - Registered Nurses and Miidwives"
    ],
    "CHPPD – Healthcare Support Workers": [
        "CHPPD – Healthcare Support Workers",
        "CHPPD - Healthcare Support Workers"
    ],
    "CHPPD -  Registered Nursing Associates": [
        "CHPPD -  Registered Nursing Associates",
        "CHPPD - Registered Nursing Associates"
    ],
    "CHPPD - Unregistered Trainee Nursing Associates": [
        "CHPPD - Unregistered Trainee Nursing Associates",
        "CHPPD - Unregisterd Trainee Nursing Associates"
    ],
}

for final_name, candidates in column_groups.items():
    # só usa as colunas que realmente existem no df
    existing = [c for c in candidates if c in combined_data.columns]
    if not existing:
        continue  # não há nenhuma destas colunas no df, passa à frente

    # combina: usa o primeiro valor não nulo da lista de colunas
    combined_data[final_name] = combined_data[existing].bfill(axis=1).iloc[:, 0]

    # apaga as colunas antigas com nomes “errados”, mantendo só a final
    cols_to_drop = [c for c in existing if c != final_name]
    combined_data.drop(columns=cols_to_drop, inplace=True)

```
