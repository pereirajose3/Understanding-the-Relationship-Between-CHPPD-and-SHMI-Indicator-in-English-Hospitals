### Fill Missing or Zero CHPPD Values Using Previous Month for Each Organisation

```python
import pandas as pd
import numpy as np

# trabalhar directamente no df result
df = result.copy()

# 🔹 NÃO converter year_month para datetime – usar como string
# Garantir só que não há espaços estranhos
df["year_month"] = df["year_month"].astype(str).str.strip()

# ordenar por organização e ano/mês (como string "YYYY-MM" funciona bem)
df = df.sort_values(["Organisation Name_x", "year_month"])

# colunas a corrigir
cols_to_fix = [
    "CHPPD - Overall",
    "CHPPD – Registered Nurses and Midwives",
    "Non-registered Nurses"
]

# substituir zeros por NaN
df[cols_to_fix] = df[cols_to_fix].replace(0, np.nan)

# forward fill dentro de cada organização
df[cols_to_fix] = df.groupby("Organisation Name_x")[cols_to_fix].ffill()

# backfill se o primeiro registo da org for NaN
df[cols_to_fix] = df.groupby("Organisation Name_x")[cols_to_fix].bfill()

# guardar de volta
result = df.copy()

```
